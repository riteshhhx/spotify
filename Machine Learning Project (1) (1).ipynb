# MACHINE LEARNING MODEL

```python
from sklearn.model_selection import train_test_split   # Split data into train/test
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression      # Regression model
from sklearn.linear_model import LogisticRegression    # Classification model

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    confusion_matrix,
    classification_report
)

from sklearn.metrics.pairwise import cosine_similarity # Recommendation system
```

```python
numeric_features = [
    column for column in data.columns
    if data[column].dtype.kind in "iuf"    # i = int, u = unsigned int, f = float
]
```

```python
# Target column for regression
target_score = "Track Score"
```

```python

# Input feature columns (exclude the target)
features = [column for column in numeric_features if column != target_score]
```

```python
# Prepare feature matrix (X) and target vector (y)
feature_data = data[features]        # All input features
target_values = data[target_score]   # Track Score values

print("\nNumber of input features:", len(features))
print("Sample feature columns:", features[:10]) 
```

# LINEAR REGRESSION MODEL
# (TRACK SCORE PREDICTION)

```python
# Split data into training and testing sets for regression
train_features, test_features, train_target, test_target = train_test_split(
    feature_data,           # Input features (X)
    target_values,          # Target values (y)
    test_size=0.2,          # 20% of data used for testing
    random_state=42         # Random seed for reproducibility
)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(train_features, train_target)
```

```python
predicted_scores = linear_model.predict(test_features)
```

```python
mae = mean_absolute_error(test_target, predicted_scores)                   # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(test_target, predicted_scores))          # Root Mean Squared Error
r2 = r2_score(test_target, predicted_scores)                               # R-squared value
```

```python
print("\nLinear Regression Results:")
print("MAE  (Mean Absolute Error):", mae)
print("RMSE (Root Mean Squared Error):", rmse)
print("R²   (R-squared):", r2)
```

```python
plt.scatter(test_target, predicted_scores, alpha=0.6)
plt.xlabel("Actual Track Score")
plt.ylabel("Predicted Track Score")
plt.title("Actual vs Predicted Track Score")
plt.plot(
    [test_target.min(), test_target.max()],
    [test_target.min(), test_target.max()],
    color='red'
)
plt.show()
```

```python
# Pick one sample song from test set
sample_song_features = test_features.iloc[0]

# Predict its Track Score
sample_prediction = linear_model.predict([sample_song_features])

print("Actual Track Score :", test_target.iloc[0])
print("Predicted Track Score:", sample_prediction[0])
```

# SONG RECOMMENDATION SYSTEM (CONTENT-BASED)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

scaler = StandardScaler()
feature_data_scaled = scaler.fit_transform(feature_data)

# Calculate similarity between all songs using scaled features
similarity_matrix = cosine_similarity(feature_data_scaled)
```

```python
similarity_matrix.shape
```

```python
def recommend_songs(song_name, top_n=5):
    
    
    # Find the selected song (case-insensitive)
    selected_song = data[data['Track'].str.lower() == song_name.lower()]
    
    if selected_song.empty:
        print("Song not found in dataset")
        return
    
    # Get index of the selected song
    selected_index = selected_song.index[0]
    
    # Get similarity scores
    similarity_scores = similarity_matrix[selected_index]
    
    # Pair index with similarity score
    song_scores = list(enumerate(similarity_scores))
    
    # Sort by similarity (highest first)
    song_scores = sorted(song_scores, key=lambda x: x[1], reverse=True)
    
    print(f"\nYou selected: {data.loc[selected_index, 'Track']} - {data.loc[selected_index, 'Artist']}")
    print("Recommended songs:\n")
    
    # Skip first entry (same song)
    for i in range(1, top_n + 1):
        song_index = song_scores[i][0]
        score = song_scores[i][1]
        
        print(
            f"{data.loc[song_index, 'Track']} - {data.loc[song_index, 'Artist']} "
            f"(Similarity Score: {score:.3f})"
        )
```

```python
sample_song = data['Track'].iloc[0]
recommend_songs(sample_song, 5)
```
